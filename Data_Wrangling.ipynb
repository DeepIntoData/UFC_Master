```python
# Dependencies
import pandas as pd
import csv

#Set CSV to a Variable
MASTER = "Resources/ufc_master_data.csv"
data = pd.read_csv(MASTER)
data
```

```python
#Goal is to split the data from R and B fighters to a winner and loser table with PK set to match IDs
#Create Match IDs

data.insert(0, 'Match ID', range(1, 1 + len(data)))
```

```python
#Create a List and View Column Names - Used to Drop/Rename columns later
column_names = []

for col_name in data.columns: 
    column_names.append(col_name)
    print(col_name)
    
```

```python
#Drop irrelevant columns - We drop the Win_Name because it has mostly NaN values. Instead we will choose 'Winner' column

data_df = data.drop(columns=['Win_Name'])
```

```python
#CREATE WINNERS_DF
#Sort by Blue Winners/Red Winners
B_Win_df = data_df.loc[data['Winner'] == 'Blue']
R_Win_df = data_df.loc[data['Winner'] == 'Red']

#Drop loser columns
Blue=B_Win_df.drop(columns=['R_fighter'])
Red=R_Win_df.drop(columns=['B_fighter'])

#Rename Column
Blue_df = Blue.rename(columns={"B_fighter":"Name"})
Red_df = Red.rename(columns={"R_fighter":"Name"})

#Concat
Winners_df = pd.concat([Blue_df,Red_df])

#Sort and display
Winners = Winners_df.sort_values(by='Match ID', ascending=True) 
Winners
```

```python
#Create Losers_df

B_Win_df = data_df.loc[data['Winner'] == 'Blue']
R_Win_df = data_df.loc[data['Winner'] == 'Red']

#Drop Winner columns
Blue_L=B_Win_df.drop(columns=['B_fighter'])
Red_L=R_Win_df.drop(columns=['R_fighter'])

#Rename Column
Blue_L_df = Blue_L.rename(columns={"R_fighter":"Name"})
Red_L_df = Red_L.rename(columns={"B_fighter":"Name"})

#Concat
Losers_df = pd.concat([Blue_L_df,Red_L_df])

#Sort and display
Losers = Losers_df.sort_values(by='Match ID', ascending=True) 
Losers
```

```python
#Save to CSV

Winners.to_csv('UFC_Winners.csv', encoding='utf-8')
Losers.to_csv('UFC_Losers.csv', encoding='utf-8')
```
